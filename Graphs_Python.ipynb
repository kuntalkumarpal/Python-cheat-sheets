### Vertical Bar with two towers

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['Abductive NLI', 'Physical IQA', 'Social IQA']
men_means = [20, 48, 30]
women_means = [25, 52, 34]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Good K')
rects2 = ax.bar(x + width/2, women_means, width, label='Insufficient K')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('# of Samples')
ax.set_title('Correct Prediction Analysis')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

```

### Vertical Bar graph with 4 towers

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['Abductive NLI', 'Physical IQA', 'Social IQA']
noK = [0, 48, 36]
correctOptK = [39, 15, 35]
exactK = [62,52,0]
relatedK = [24,0,29]


x = np.arange(len(labels))  # the label locations
width = 0.21  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - 3*width/2, noK, width, label='No K')
rects2 = ax.bar(x + width/2, correctOptK, width, label='K across True Option')
rects1 = ax.bar(x - width/2, exactK, width, label='Exact K')
rects2 = ax.bar(x + 3*width/2, relatedK, width, label='Related K')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('# of Samples')
ax.set_title('Correct Prediction Analysis')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.set_ylim([0, 80])
```
