# Merge Pandas dataframes

```python
zoo_eats = pd.DataFrame([['elephant','vegetables'], ['tiger','meat'], ['kangaroo','vegetables'], ['zebra','vegetables'], ['giraffe','vegetables']], columns=['animal', 'food'])
```

```python
zoo_eats
```

```python
zoo.merge(zoo_eats) #inner join
```

```python
zoo.merge(zoo_eats, how='outer')
```

```python
zoo.merge(zoo_eats,how='left')
```

```python
zoo.merge(zoo_eats,how='left',left_on='animal', right_on = 'animal') # Specifying the col name for join
```

### Sorting

```python
zoo.sort_values('water_need').head()
```

```python
zoo.sort_values(['animal','water_need'])
zoo.sort_values(by = ['animal', 'water_need']).head()
```

```python
zoo.sort_values(by = ['water_need'],ascending=False)
```

```python
zoo.sort_values(by = ['water_need'],ascending=False).reset_index()
# resets the index after any manipulation have been done on the dataframe otherwise there may be issues with visualizations or matplotlibs
```

```python
zoo.sort_values(by = ['water_need'],ascending=False).reset_index(drop=True)
# Drops the old indices as well
```
